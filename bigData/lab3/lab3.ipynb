<p>5. Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.</p>

```python
plt.figure(figsize=(10, 5))

plt.title('Графики в одной координатной плоскости')

plt.xlabel('Ось x', labelpad=20)

plt.axhline(0, color='black')
plt.axvline(0, color='black')

x = np.arange(-7, 11, 0.1)

y1 = (x - 2)**2 - 1

legend1 = '(x-2)^2 - 1'

y2 = np.sin(x) * np.cos(x)
legend2 = 'sin(x) * cos(x)'

ax_01 = plt.gca()

graph1, = ax_01.plot(x, y1, 'b-')

ax_01.axis([-4, 6,   -4, 6])
ax_01.grid(color='r')
ax_01.set_ylabel(legend1, color = 'b', labelpad=20)

ax_02 = ax_01.twinx()

graph2, = ax_02.plot(x, y2, 'r-')

ax_02.axis([-8, 8, -8, 8])
ax_02.grid(color = 'b')
ax_02.set_ylabel(legend2, color='r', labelpad=20)

ax_02.legend((graph1, graph2), (legend1, legend2), loc='best')

plt.show()
```

<P>6. Построить график функции, заданной параметрически, xt=cos 3t, y=sin 2t, t∈[0,4π]</p>

```python
plt.title('Параметрический график x=cos 3t, y=sin 4t, t∈[0,4π]')

t = np.linspace(0, 4 * np.pi, 100)
x = np.cos(3 * t)
y = np.sin(4 * t)

axx = plt.subplot(1,1,1)
axx.set_aspect(1)

plt.xticks([i for i in np.linspace(-1.1, 1.1, 3)])
plt.yticks([i for i in np.linspace(-1.1, 1.1, 10)])

plt.plot(x, y)

plt.grid(True)
```

<p>7. Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице</p>

```python
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.3]
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

plt.figure(figsize=(12, 6))
plt.bar(years, values, width=0.6, color='skyblue', edgecolor='black', alpha=0.7, label='Прожиточный минимум')

plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('График прожиточного минимума в КР (2012-2022)')
plt.xticks(years)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

<p>8. Построить круговой график по данным п.7</p>

```python
plt.figure(figsize=(8, 8))
plt.title('Круговой график прожиточного минимума в КР (2012-2022)')

plt.pie(values, labels=years, autopct='%.1f%%', startangle=140)

plt.axis('equal') 
plt.show()
```

<p>9. Построить график произвольной каркасной поверхности.</p>

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

u, v = np.meshgrid(u, v)

# Определение функции, задающей поверхность
def custom_surface(u, v):
    x = np.cos(u) * np.sin(v)
    y = np.sin(u) * np.sin(v)
    z = np.cos(v)
    return x, y, z

x, y, z = custom_surface(u, v)

figure = plt.figure()

ax = figure.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

plt.title('Произвольная каркасная поверхность')
plt.show()
```
